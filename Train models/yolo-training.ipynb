```python
# library importing 
import torch
import torchvision
import ultralytics
from ultralytics import YOLO
```

```python
# setting up GPU usage
# check if GPU is usable and available
print("GPU usable?", torch.cuda.is_available())
print("Number of GPU available on device:", torch.cuda.device_count())

# set runtime to GPU
device = "0" if torch.cuda.is_available() else "cpu"
if device == "0":
    # get GPU information
    print("Device:", torch.cuda.device(0))
    print("Device name:", torch.cuda.get_device_name(0))
    
    # set GPU
    torch.cuda.set_device(0)

# end
print('---Done setting up!---')
```

```python
# perform setting GPU runtime on Yolov8
# https://stackoverflow.com/questions/75691440/how-to-move-yolov8-model-onto-gpu

# load Yolov8 model
yolo_model = YOLO('/kaggle/input/pokemon-dataset/last.pt')
print('Device Yolov8 was using before set:', yolo_model.device.type)

# move yolo model onto GPU
yolo_model.to('cuda')

# if you done things right, yolo model now is shifted onto GPU
print('Current Yolov8 device:', yolo_model.device.type)
```

```python
# start training model with custom dataset

"""
Hyper parameters include:
- data: path to dataset description
- epochs: number of training epochs
- device: set training on GPU NVIDIA GeForce GTX 1650, device number 0, as seen above
"""

yolo_model.train(data='/kaggle/input/pokemon-dataset/train_kaggle.yaml', epochs=300, device=[0,1], save_period=50)
```
